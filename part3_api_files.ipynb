```python
# -------------------------------
# Part A — Write & Append
# -------------------------------

file_name = "python_notes.txt"

# Step 1: Write initial 5 lines
try:
    with open(file_name, "w", encoding="utf-8") as file:
        file.write("Topic 1: Variables store data. Python is dynamically typed.\n")
        file.write("Topic 2: Lists are ordered and mutable.\n")
        file.write("Topic 3: Dictionaries store key-value pairs.\n")
        file.write("Topic 4: Loops automate repetitive tasks.\n")
        file.write("Topic 5: Exception handling prevents crashes.\n")
    print("File written successfully.")
except Exception as e:
    print("Error while writing file:", e)


# Step 2: Append 2 more lines
try:
    with open(file_name, "a", encoding="utf-8") as file:
        file.write("Topic 6: Functions help reuse code.\n")
        file.write("Topic 7: Classes enable object-oriented programming.\n")
    print("Lines appended.")
except Exception as e:
    print("Error while appending:", e)


# -------------------------------
# Part B — Read & Process
# -------------------------------

try:
    with open(file_name, "r", encoding="utf-8") as file:
        lines = file.readlines()

    print("\n--- File Content (Numbered) ---")

    # Step 1: Print numbered lines
    for i, line in enumerate(lines, start=1):
        print(f"{i}. {line.strip()}")

    # Step 2: Count total lines
    print(f"\nTotal lines: {len(lines)}")

    # Step 3: Keyword search
    keyword = input("\nEnter keyword to search: ").lower()

    found = False
    print("\nMatching lines:")

    for line in lines:
        if keyword in line.lower():
            print(line.strip())
            found = True

    if not found:
        print("No matching lines found.")

except Exception as e:
    print("Error while reading file:", e)
```

```python
# If requests not installed
!pip install requests

import requests

BASE_URL = "https://dummyjson.com/products"

# -------------------------------
# Step 1 — Fetch & Display Products
# -------------------------------
try:
    response = requests.get(f"{BASE_URL}?limit=20")
    response.raise_for_status()  # error handle

    data = response.json()
    products = data["products"]

    print("\n--- Product List ---")
    print(f"{'ID':<5} {'Title':<25} {'Category':<15} {'Price':<10} {'Rating'}")
    print("-" * 70)

    for p in products:
        print(f"{p['id']:<5} {p['title'][:24]:<25} {p['category']:<15} ${p['price']:<9} {p['rating']}")

except Exception as e:
    print("Error fetching products:", e)


# -------------------------------
# Step 2 — Filter & Sort
# -------------------------------
try:
    filtered = [p for p in products if p["rating"] >= 4.5]

    sorted_products = sorted(filtered, key=lambda x: x["price"], reverse=True)

    print("\n--- Filtered (rating >= 4.5) & Sorted by Price ---")

    for p in sorted_products:
        print(f"{p['title']} | Price: ${p['price']} | Rating: {p['rating']}")

except Exception as e:
    print("Error in filtering/sorting:", e)


# -------------------------------
# Step 3 — Search by Category (laptops)
# -------------------------------
try:
    response = requests.get(f"{BASE_URL}/category/laptops")
    response.raise_for_status()

    laptops = response.json()["products"]

    print("\n--- Laptops Category ---")

    for p in laptops:
        print(f"{p['title']} | Price: ${p['price']}")

except Exception as e:
    print("Error fetching laptops:", e)


# -------------------------------
# Step 4 — POST Request (Simulated)
# -------------------------------
try:
    new_product = {
        "title": "My Custom Product",
        "price": 999,
        "category": "electronics",
        "description": "A product I created via API"
    }

    response = requests.post(f"{BASE_URL}/add", json=new_product)
    response.raise_for_status()

    print("\n--- POST Response ---")
    print(response.json())

except Exception as e:
    print("Error in POST request:", e)
```

```python
import requests
from datetime import datetime

# -------------------------------
# Helper: Error Logger
# -------------------------------
def log_error(message):
    with open("error_log.txt", "a") as f:
        f.write(f"{datetime.now()} - {message}\n")


# -------------------------------
# Part A — Guarded Calculator
# -------------------------------
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"


print("\n--- Safe Divide Tests ---")
print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))


# -------------------------------
# Part B — Guarded File Reader
# -------------------------------
def read_file_safe(filename):
    try:
        with open(filename, "r") as f:
            return f.read()
    except FileNotFoundError:
        msg = f"Error: File '{filename}' not found."
        log_error(msg)
        print(msg)
    finally:
        print("File operation attempt complete.")


print("\n--- File Reader Tests ---")
print(read_file_safe("python_notes.txt"))   # should work
print(read_file_safe("ghost_file.txt"))     # should fail


# -------------------------------
# Part C — Robust API Calls
# -------------------------------
BASE_URL = "https://dummyjson.com/products"

def fetch_products():
    try:
        res = requests.get(f"{BASE_URL}?limit=5", timeout=5)
        res.raise_for_status()
        return res.json()["products"]
    except requests.exceptions.ConnectionError:
        msg = "Connection failed. Please check your internet."
        print(msg)
        log_error(msg)
    except requests.exceptions.Timeout:
        msg = "Request timed out. Try again later."
        print(msg)
        log_error(msg)
    except Exception as e:
        print("Error:", e)
        log_error(str(e))


print("\n--- Robust API Test ---")
products = fetch_products()
if products:
    for p in products:
        print(p["title"], "-", p["price"])


# -------------------------------
# Part D — Input Validation Loop
# -------------------------------
print("\n--- Product Lookup ---")

while True:
    user_input = input("Enter product ID (1-100) or 'quit': ")

    if user_input.lower() == "quit":
        print("Exiting...")
        break

    # validation
    if not user_input.isdigit():
        print("Invalid input. Enter a number.")
        continue

    product_id = int(user_input)

    if not (1 <= product_id <= 100):
        print("Enter ID between 1 and 100.")
        continue

    # API call
    try:
        res = requests.get(f"{BASE_URL}/{product_id}", timeout=5)

        if res.status_code == 404:
            print("Product not found.")
        elif res.status_code == 200:
            data = res.json()
            print(f"{data['title']} | Price: ${data['price']}")
        else:
            print("Unexpected response.")

    except requests.exceptions.ConnectionError:
        msg = "Connection failed."
        print(msg)
        log_error(msg)
    except requests.exceptions.Timeout:
        msg = "Request timed out."
        print(msg)
        log_error(msg)
    except Exception as e:
        print("Error:", e)
        log_error(str(e))
```

```python
from datetime import datetime
import requests

LOG_FILE = "error_log.txt"

# -------------------------------
# Logger Function (proper format)
# -------------------------------
def log_error(source, message):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    with open(LOG_FILE, "a") as f:
        f.write(f"[{timestamp}] ERROR in {source}: {message}\n")


# -------------------------------
# Case 1 — Connection Error
# -------------------------------
def test_connection_error():
    try:
        requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)
    except requests.exceptions.ConnectionError:
        msg = "ConnectionError — No connection could be made"
        print(msg)
        log_error("fetch_products", msg)


# -------------------------------
# Case 2 — HTTP 404 Error
# -------------------------------
def test_404_error():
    try:
        res = requests.get("https://dummyjson.com/products/999", timeout=5)

        if res.status_code != 200:
            msg = "HTTPError — 404 Not Found for product ID 999"
            print(msg)
            log_error("lookup_product", msg)

    except Exception as e:
        log_error("lookup_product", str(e))


# -------------------------------
# Run Tests
# -------------------------------
print("\n--- Logging Tests ---")

test_connection_error()
test_404_error()


# -------------------------------
# Print log file content
# -------------------------------
print("\n--- error_log.txt content ---")

try:
    with open(LOG_FILE, "r") as f:
        print(f.read())
except FileNotFoundError:
    print("Log file not found.")
```
